#### Bài tập 2:

#### Một phép biến đổi dữ liệu của $x_i$ và $y_i$ (như việc lấy logarit) sẽ thay đổi cách diễn giải của tham số độ dốc $\beta$.

(a) Chứng minh rằng trong mô hình hồi quy $\log(y_i) = \alpha + \beta \log(x_i) + \varepsilon_i$, độ co giãn (elasticity) của $y$ đối với $x$ bằng $\beta$ (tức là, không phụ thuộc vào giá trị của $x_i$ và $y_i$).

(b) Xác định độ co giãn của $y$ đối với $x$ trong mô hình $y_i = \alpha + \beta \log(x_i) + \varepsilon_i$.

(c) Xác định độ co giãn của $y$ đối với $x$ trong mô hình $\log(y_i) = \alpha + \beta x_i + \varepsilon_i$.

**Lưu ý**:
Như trong phần lớn các môn học, `log` viết tắt cho `logarit` tự nhiên, với cơ số $e \approx 2.71828$. Trên máy tính, thường được viết là $\ln(x)$, nhưng chúng ta sử dụng ký hiệu $\log(x)$. Đạo hàm của $\log(x)$, với cơ số $e$, là $1/x$.

#### Công thức tính độ co giãn:
$$
\text{elasticity} = \frac{\partial y_i}{\partial x_i} \times \frac{x_i}{y_i}
$$
Trong đó:
- $\frac{\partial y_i}{\partial x_i}$ là đạo hàm của $y_i$ theo $x_i$, biểu thị sự thay đổi của $y$ khi $x$ thay đổi.
- $\frac{x_i}{y_i}$ là tỷ lệ của $x_i$ với $y_i$.

#### Ví dụ về độ co giãn:
- $x_i = 20$ (nghìn đồng), $y_i = 500$ (triệu đồng)
- $\frac{\partial y_i}{\partial x_i} = -10$ (doanh thu giảm 10 triệu đồng khi giá tăng 1 nghìn đồng)

Tính độ co giãn:

$$
\text{elasticity} = \frac{\partial y_i}{\partial x_i} \times \frac{x_i}{y_i} = -10 \times \frac{20}{500} = -0.4
$$

Kết quả: Độ co giãn là $-0.4$, nghĩa là khi giá tăng 1%, doanh thu giảm 0.4%.

(a) **Chứng minh rằng trong mô hình hồi quy $\log(y_i) = \alpha + \beta \log(x_i) + \varepsilon_i$, độ co giãn (elasticity) của $y$ đối với $x$ bằng $\beta$:**


$$
y_i = e^{\alpha + \beta \log(x_i) + \varepsilon_i} = e^{\alpha + \varepsilon_i} \cdot x_i^{\beta}  
$$

Đạo hàm của $y$ theo $x$:

$$
\frac{\partial y_i}{\partial x_i} = \frac{\partial}{\partial x_i} \left( e^{\alpha + \varepsilon_i} \cdot x_i^{\beta} \right) = e^{\alpha + \varepsilon_i} \cdot \beta x_i^{\beta-1} 
$$

Thay vào biểu thức độ co giãn:

$$
\text{Elasticity} = \frac{\partial y_i}{\partial x_i} \cdot \frac{x_i}{y_i} = \frac{e^{\alpha + \varepsilon_i} \cdot \beta x_i^{\beta-1} }{e^{\alpha + \varepsilon_i} \cdot x_i^{\beta} } \cdot x_i = \beta
$$

Do đó, độ co giãn của $y$ đối với $x$ trong mô hình này là $\beta$, và không phụ thuộc vào giá trị của $x_i$ hay $y_i$.

---

(b) **Xác định độ co giãn của $y$ đối với $x$ trong mô hình $y_i = \alpha + \beta \log(x_i) + \varepsilon_i$:**


Đạo hàm của $y_i$ theo $x_i$:

$$
\frac{\partial y_i}{\partial x_i} = \frac{\beta}{x_i}
$$

Độ co giãn được tính là:

$$
\text{Elasticity} = \frac{\partial y_i}{\partial x_i} \cdot \frac{x_i}{y_i} = \frac{\beta}{x_i} \cdot \frac{x_i}{y_i} = \frac{\beta}{y_i}
$$

Vậy độ co giãn trong trường hợp này là $\frac{\beta}{y_i}$.

---

(c) **Xác định độ co giãn của $y$ đối với $x$ trong mô hình $\log(y_i) = \alpha + \beta x_i + \varepsilon_i$:**


$$
y_i = e^{\alpha + \beta x_i + \varepsilon_i} = e^{\alpha + \varepsilon_i} \cdot e^{\beta x_i}
$$

Đạo hàm của $y_i$ theo $x_i$ là:

$$
\frac{\partial y_i}{\partial x_i} = e^{\alpha + \varepsilon_i} \cdot \beta e^{\beta x_i} = \beta y_i
$$

Độ co giãn là:

$$
\text{Elasticity} = \frac{\partial y_i}{\partial x_i} \cdot \frac{x_i}{y_i} = \beta y_i \cdot \frac{x_i}{y_i} = \beta x_i
$$

Do đó, độ co giãn của $y$ đối với $x$ trong mô hình này là $\beta x_i$.

```python
# Thời gian về đích ở nội dung điền kinh 100 mét dành cho nam tại Thế vận hội Olympic từ 1948-2004.
# Game: số thứ tự của kỳ Thế vận hội (1-15)
# Year: năm diễn ra Thế vận hội Olympic (1948-2004)
# Winning time men: thời gian về đích ở nội dung điền kinh 100 mét dành cho nam (tính bằng giây)
df_2 = pd.DataFrame({
    "Game": list(range(1, 16)),
    "Year": [1948, 1952, 1956, 1960, 1964, 1968, 1972, 1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004],
    "Winning time men": [10.3, 10.4, 10.5, 10.2, 10.0, 9.95, 10.14, 10.06, 10.25, 9.99, 9.92, 9.96, 9.84, 9.87, 9.85]
})
print(df_2.to_string(index=False))
```

#### Bài tập 3:

#### Bộ dữ liệu `df_2` chứa thời gian về đích ở nội dung điền kinh 100 mét dành cho nam tại Thế vận hội Olympic từ năm 1948 đến 2004. Các năm từ 1948-2004 được chuyển đổi thành số thứ tự của kỳ Thế vận hội (Game) từ 1 đến 15 để đơn giản hóa việc tính toán. Một mô hình hồi quy đơn giản cho xu hướng trong thời gian về đích được thể hiện bằng: $W_i = \alpha + \beta G_i + \varepsilon_i$, trong đó $W_i$ là "thời gian về đích" và $G_i$ là số thứ tự của kỳ Thế vận hội.

(a) Tính toán $a$ và $b$ (hệ số hồi quy), và xác định giá trị của $R^2$ và $s$ (độ lệch chuẩn của sai số).

(b) Bạn có tự tin vào khả năng dự đoán của mô hình này không? Hãy giải thích lý do.

(c) Dự đoán thời gian về đích cho các năm 2008, 2012, và 2016. So sánh dự đoán của bạn với thời gian về đích thực tế nếu có.

```python
# (a)
import statsmodels.api as sm

X = df_2['Game']
y = df_2['Winning time men']

X = sm.add_constant(X)

#Mô hình Linear
model = sm.OLS(y, X)
results = model.fit()

alpha = results.params['const']
beta = results.params['Game']

R_squared = results.rsquared
s = np.sqrt(results.mse_resid)

print(f'Hệ số hồi quy alpha: {alpha}')
print(f'Hệ số hồi quy beta: {beta}')
print(f'R^2: {R_squared}')
print(f'Độ lệch chuẩn của sai số (s): {s}')
```

(b) 
Mô hình có khả năng dự đoán kém vì $R^2$ không quá cao và độ lệch chuẩn cao.

```python
# (c)
future_games = pd.DataFrame({
    'Game': [16, 17, 18]
})

future_games = sm.add_constant(future_games)

predictions = results.predict(future_games)

for i, year in enumerate([2008, 2012, 2016]):
    print(f'Dự đoán thời gian về đích cho năm {year}: {predictions[i]} giây')
```

So sánh với thời gian thực tế: 2008 (Usain Bolt): 9.69 giây, 2012 (Usain Bolt): 9.63 giây và 2016 (Usain Bolt): 9.81 giây.

#### Bài tập 4:

#### Xét tình huống biến số $x$ được quan sát với sai số đo lường (measurement error), điều này khá phổ biến đối với các biến kinh tế vĩ mô phức tạp như thu nhập quốc dân.

#### Giả sử $x^*$ là biến kinh tế thực sự nhưng không được quan sát, và quá trình tạo dữ liệu được cho bởi $y_i = \alpha + \beta x_i^* + \varepsilon_i^*$, trong đó $x_i^*$ và $\varepsilon_i^*$ không tương quan.

#### Các giá trị $x$ quan sát được là $x_i = x_i^* + v_i$, với sai số đo lường $v_i$ không tương quan với $x_i^*$ và $\varepsilon_i^*$. Tỷ lệ tín hiệu trên nhiễu (signal-to-noise ratio) được định nghĩa là $SN = \frac{\sigma_{x^*}^2}{\sigma_v^2}$, trong đó $\sigma_{x^*}^2$ là phương sai (variance) của $x^*$ và $\sigma_v^2$ là phương sai của $v$.

#### Mô hình hồi quy ước lượng là $y_i = \alpha + \beta x_i + \varepsilon_i$, và chúng ta xem xét ước lượng bình phương tối thiểu $b$ của $\beta$ ($b$ là ước lượng, $\beta$ là hệ số thực).

(a) Bạn có nghĩ rằng giá trị của $b$ phụ thuộc vào phương sai của sai số đo lường không? Tại sao?

(b) Chứng minh rằng $b = \beta + \frac{\sum_{i=1}^{n}(x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$.

(c) Chứng minh rằng $\varepsilon_i = \varepsilon_i^* - \beta v_i$.

(d) Chứng minh rằng hiệp phương sai (covariance) giữa $x_i$ và $\varepsilon_i$ bằng $- \beta \sigma_v^2$.

(e) Chứng minh rằng với cỡ mẫu lớn $n$, ta có $b - \beta \approx - \beta \frac{\sigma_v^2}{\sigma_{x^*}^2 + \sigma_v^2}$.

(f) Tính xấp xỉ độ chệch (bias) $b - \beta$ cho $\beta = 1$ trong các trường hợp $SN = 1$, $SN = 3$, và $SN = 10$.

### (a) Giá trị của $ b $ có phụ thuộc vào phương sai của sai số đo lường không?

Có, giá trị của $ b $ phụ thuộc vào phương sai của sai số đo lường. Điều này xảy ra vì $ x_i $ là biến quan sát được với sai số đo lường, và sai số đo lường làm giảm mối quan hệ thực sự giữa $ y_i $ và $ x_i^* $. Cụ thể, khi phương sai của sai số đo lường tăng lên (tức là $ \sigma_v^2 $ lớn hơn), tín hiệu thực $ x_i^* $ bị nhiễu nhiều hơn, dẫn đến việc ước lượng $ b $ bị "làm mờ" và sẽ bị chệch xuống so với giá trị thực $ \beta $. Tỷ lệ tín hiệu trên nhiễu $ SN $ là thước đo quan trọng để xác định mức độ ảnh hưởng của sai số đo lường.

---

### (b) Chứng minh rằng $ b = \beta + \frac{\sum_{i=1}^{n}(x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{\sum_{i=1}^{n}(x_i - \bar{x})^2} $:

Phương trình hồi quy ước lượng:

$$
y_i = \alpha + \beta x_i + \varepsilon_i
$$

Ước lượng bình phương tối thiểu $ b $ cho $ \beta $:

$$
b = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$

Biểu thức của $ y_i - \bar{y} $:

$$
y_i - \bar{y} = \beta (x_i - \bar{x}) + (\varepsilon_i - \bar{\varepsilon})
$$

Thay vào công thức của $ b $:

$$
b = \frac{\sum_{i=1}^{n}(x_i - \bar{x}) \left[ \beta (x_i - \bar{x}) + (\varepsilon_i - \bar{\varepsilon}) \right]}{\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$

Nhân phân phối:

$$
b = \frac{\sum_{i=1}^{n} \beta (x_i - \bar{x})^2 + \sum_{i=1}^{n} (x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$

Suy ra:

$$
b = \beta + \frac{\sum_{i=1}^{n}(x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$


---

### (c) Chứng minh rằng $ \varepsilon_i = \varepsilon_i^* - \beta v_i $:

Mô hình thực:

$$
y_i = \alpha + \beta x_i^* + \varepsilon_i^*
$$

Mô hình ước lượng:

$$
y_i = \alpha + \beta x_i + \varepsilon_i
$$

Từ đó, $ x_i = x_i^* + v_i $, suy ra:

$$
y_i = \alpha + \beta (x_i^* + v_i) + \varepsilon_i = \alpha + \beta x_i^* + \beta v_i + \varepsilon_i
$$

So sánh với mô hình thực $ y_i = \alpha + \beta x_i^* + \varepsilon_i^* $, ta có:

$$
\varepsilon_i = \varepsilon_i^* - \beta v_i
$$

---

### (d) Chứng minh rằng hiệp phương sai giữa $ x_i $ và $ \varepsilon_i $ bằng $ - \beta \sigma_v^2 $:

Hiệp phương sai $ \text{Cov}(x_i, \varepsilon_i) $:

$$
\text{Cov}(x_i, \varepsilon_i) = \text{Cov}(x_i^* + v_i, \varepsilon_i^* - \beta v_i)
$$

Sử dụng các tính chất của hiệp phương sai, ta có:

$$
\text{Cov}(x_i, \varepsilon_i) = \text{Cov}(x_i^*, \varepsilon_i^*) + \text{Cov}(x_i^*, -\beta v_i) + \text{Cov}(v_i, \varepsilon_i^*) + \text{Cov}(v_i, -\beta v_i)
$$

Vì $ x_i^* $ không tương quan với $ \varepsilon_i^* $ và $ v_i $, ta có:

$$
\text{Cov}(x_i^*, \varepsilon_i^*) = 0, \quad \text{Cov}(x_i^*, v_i) = 0, \quad \text{Cov}(v_i, \varepsilon_i^*) = 0
$$

Do đó:

$$
\text{Cov}(x_i, \varepsilon_i) = \text{Cov}(v_i, -\beta v_i) = -\beta \text{Var}(v_i) = -\beta \sigma_v^2
$$

---

### (e) Chứng minh rằng với cỡ mẫu lớn $ n $, ta có $ b - \beta \approx - \beta \frac{\sigma_v^2}{\sigma_{x^*}^2 + \sigma_v^2} $:

Từ phần (d), ta biết rằng:

$$
\text{Cov}(x_i, \varepsilon_i) = - \beta \sigma_v^2
$$

Sử dụng kết quả từ phần (b), ước lượng $ b $ có thể viết là:

$$
b = \beta + \frac{\sum_{i=1}^{n}(x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$

Khi $ n $ lớn, $ \frac{\sum_{i=1}^{n}(x_i - \bar{x})(\varepsilon_i - \bar{\varepsilon})}{n} $ tiến tới hiệp phương sai $ \text{Cov}(x_i, \varepsilon_i) $, và $ \frac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n} $ tiến tới phương sai của $ x_i $, tức là $ \sigma_{x}^2 = \sigma_{x^*}^2 + \sigma_v^2 $. Do đó:

$$
b \approx \beta + \frac{-\beta \sigma_v^2}{\sigma_{x^*}^2 + \sigma_v^2}
$$

Suy ra:

$$
b - \beta \approx - \beta \frac{\sigma_v^2}{\sigma_{x^*}^2 + \sigma_v^2}
$$

---

### (f) Tính xấp xỉ độ chệch $ b - \beta $ cho $ \beta = 1 $ với $ SN = 1 $, $ SN = 3 $, và $ SN = 10 $:

Từ công thức trong phần (e), ta có:

$$
b - \beta \approx - \beta \frac{\sigma_v^2}{\sigma_{x^*}^2 + \sigma_v^2}
$$

Tỷ lệ tín hiệu trên nhiễu $ SN = \frac{\sigma_{x^*}^2}{\sigma_v^2} $, do đó $ \frac{\sigma_v^2}{\sigma_{x^*}^2} = \frac{1}{SN} $. Khi đó:

$$
b - \beta \approx - \beta \frac{1}{1 + SN}
$$

Với $ \beta = 1 $, ta có:

- **$ SN = 1 $**:

$$
b - 1 \approx - \frac{1}{1 + 1} = -0.5
$$

- **$ SN = 3 $**:

$$
b - 1 \approx - \frac{1}{1 + 3} = -0.25
$$

- **$ SN = 10 $**:

$$
b - 1 \approx - \frac{1}{1 + 10} = -0.0909
$$

Kết quả cho thấy độ chệch giảm khi $ SN $ tăng.

```python
# Thời gian về đích ở nội dung điền kinh 100 mét dành cho nam và nữ tại Thế vận hội Olympic từ 1948-2004.
# Game: số thứ tự của kỳ Thế vận hội (1-15)
# Year: năm diễn ra Thế vận hội Olympic (1948-2004)
# Winning time men: thời gian về đích ở nội dung điền kinh 100 mét dành cho nam (tính bằng giây)
# Winning time women: thời gian về đích ở nội dung điền kinh 100 mét dành cho nữ (tính bằng giây)
df_3 = pd.DataFrame({
    "Game": list(range(1, 16)),
    "Year": [1948, 1952, 1956, 1960, 1964, 1968, 1972, 1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004],
    "Winning time men": [10.3, 10.4, 10.5, 10.2, 10.0, 9.95, 10.14, 10.06, 10.25, 9.99, 9.92, 9.96, 9.84, 9.87, 9.85],
    "Winning time women": [11.9, 11.5, 11.5, 11.0, 11.4, 11.08, 11.07, 11.08, 11.06, 10.97, 10.54, 10.82, 10.94, 10.75, 10.93]
})
print(df_3.to_string(index=False))
```

#### Bài tập 5:

#### Bộ dữ liệu `df_3` chứa thời gian về đích ở nội dung điền kinh 100 mét dành cho nam và nữ tại Thế vận hội Olympic từ năm 1948 đến 2004. Dữ liệu bao gồm thời gian về đích của các vận động viên qua các kỳ Thế vận hội (1-15) trong giai đoạn này.

(a) Sử dụng một mô hình hồi quy tuyến tính với công thức $y_{\text{men}} = a_1 \times \text{Year} + b_1$ cho nam và $y_{\text{women}} = a_2 \times \text{Year} + b_2$ cho nữ. Dự đoán khi nào thời gian về đích của nam và nữ sẽ bằng nhau. Chứng minh rằng mô hình dự đoán thời gian về đích bằng nhau vào khoảng năm 2140.

(b) Sử dụng một mô hình hồi quy phi tuyến với hàm `log` theo công thức $\log(y_{\text{men}}) = a_3 \times \text{Year} + b_3$ cho nam và $\log(y_{\text{women}}) = a_4 \times \text{Year} + b_4$ cho nữ. Dự đoán khi nào thời gian về đích của nam và nữ sẽ bằng nhau. Chứng minh rằng mô hình này dự đoán thời gian bằng nhau vào khoảng năm 2192.

(c) Chứng minh rằng mô hình hồi quy tuyến tính dự đoán thời gian về đích bằng nhau ở mức xấp xỉ 8.53 giây.

(d) Đưa ra nhận xét về những kết quả này và mô hình hồi quy được sử dụng. Bạn có nghĩ rằng dự đoán của các mô hình này đáng tin cậy cho tương lai xa không? Tại sao?

```python
import statsmodels.api as sm

X = sm.add_constant(df_3['Year']) 

# Hồi quy tuyến tính cho nam
model_men = sm.OLS(df_3['Winning time men'], X).fit()

# Hồi quy tuyến tính cho nữ
model_women = sm.OLS(df_3['Winning time women'], X).fit()

# Dự đoán năm mà thời gian về đích của nam và nữ bằng nhau
b1, a1 = model_men.params
b2, a2 = model_women.params

# Giải phương trình a1*Year + b1 = a2*Year + b2
year_equal = (b2 - b1) / (a1 - a2)

# Tính thời gian về đích tại năm bằng nhau
time_equal = a1 * year_equal + b1

# Mô hình hồi quy phi tuyến (log)
log_men = np.log(df_3['Winning time men'])
log_women = np.log(df_3['Winning time women'])

# Hồi quy logarit cho nam
model_log_men = sm.OLS(log_men, X).fit()

# Hồi quy logarit cho nữ
model_log_women = sm.OLS(log_women, X).fit()

# Tìm năm mà log(time men) = log(time women)
b3, a3 = model_log_men.params
b4, a4 = model_log_women.params

# Giải phương trình a3*Year + b3 = a4*Year + b4
year_log_equal = (b4 - b3) / (a3 - a4)

# Kết quả hồi quy tuyến tính và log
results = {
    "a1 (men)": a1,
    "b1 (men)": b1,
    "a2 (women)": a2,
    "b2 (women)": b2,
    "Năm về đích bằng nhau (linear)": year_equal,
    "Thời gian về đích bằng nhau (linear)": time_equal,
    "a3 (log men)": a3,
    "b3 (log men)": b3,
    "a4 (log women)": a4,
    "b4 (log women)": b4,
    "Năm về đích bằng nhau (log)": year_log_equal
}

results
```
